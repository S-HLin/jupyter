```python
# import keras
import keras
# import keras_retinanet
from keras_retinanet import models
from keras_retinanet.utils.image import read_image_bgr, preprocess_image, resize_image
from keras_retinanet.utils.visualization import draw_box #, draw_caption
from keras_retinanet.utils.colors import label_color
# import miscellaneous modules
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import time
from time import strftime
# set tf backend to allow memory to grow, instead of claiming everything
import tensorflow as tf
# image process
import io
import csv
from PIL import ImageFont, ImageDraw, Image
# flask server
from flask import Flask,jsonify
from flask import request
import base64

from keras.backend.tensorflow_backend import set_session

config = tf.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.3
set_session(tf.Session(config=config))
```

```python
class testmodel:   
    def __init__(self):
#         self.model_path= os.path.join('..', 'snapshots', 'resnet50_csv_trans.h5')
        self.model_path= '/dataDisk/myfloder/jupyter/snapshots/0925/inf_0925.h5'
        self.csv_path='/dataDisk/myfloder/inputdata/pascal/0925-pascal/classes.csv'
        self.isplot=True     
        self.model=self.loadmodell()     
        self.labelsname=self.csvlabell()
        
    def loadmodell(self):
        # load retinanet model
        global model
        model = models.load_model(self.model_path, backbone_name='resnet50')
        global graph
        graph=tf.get_default_graph()
        return model
      
    def csvlabell(self):
        with open(self.csv_path, newline='') as csvfile: 
          # 讀取 CSV 檔內容，將每一列轉成一個 dictionary
            reader = csv.reader(csvfile)
            labels_to_names = {}
          # 以迴圈輸出指定欄位
            for row in reader:
                labels_to_names[int(row[1])]= row[0]
          #print(labels_to_names)
        return(labels_to_names)  
    
#     def stringToRGB(self,base64_string):
#         #imgdata = base64.b64decode(str(base64_string))
#         imgdata = base64.b64decode(str(base64_string))
#         image = Image.open(io.BytesIO(imgdata))
#         return cv2.cvtColor(np.array(image), cv2.COLOR_BGR2RGB)
    
#     def base64_cv2(self,base64_str):
#         imgString = base64.b64decode(base64_str)
#         nparr = np.fromstring(imgString,np.uint8)  
#         image = cv2.imdecode(nparr,cv2.IMREAD_COLOR)
#         return image
    def base64_to_image(self,base64_code): 
        # base64解码
        img_data = base64.b64decode(base64_code)
        # 转换为np数组
        img_array = np.fromstring(img_data, np.uint8)
        # 转换成opencv可用格式
        img = cv2.imdecode(img_array, cv2.COLOR_RGB2BGR) 
        return img

    def draw_box2(self,image, box, color): 
        if self.isplot:
            draw_box(image, box, color=color)
        
    def draw_caption2(self,image,box,caption):
        if self.isplot:
            # 字體設定
            fontPath = "/usr/local/lib/python3.5/dist-packages/matplotlib/mpl-data/fonts/ttf/msj.ttf"
            font = ImageFont.truetype(fontPath,35)
            # 將 NumPy 陣列轉為 PIL 影像
            imgPil = Image.fromarray(image)  
            # 在圖片上加入文字
            draw = ImageDraw.Draw(imgPil)
            draw.text((box[0], box[1] - 10), caption, font = font, fill = (255,255,255))
            # 將 PIL 影像轉回 NumPy 陣列
            image = np.array(imgPil)
        return image
    
    def plotsubfig(self,img,fig,x,y,n,size): 
        if self.isplot:
            plt.figure(fig,figsize=(size,size))
            plt.subplot(x,y,n) # 2,4
            plt.axis('off')
            plt.imshow(img)

    def plotfig(self,img,fig,size): 
        if self.isplot:
            plt.figure(fig,figsize=(size,size))
            plt.axis('off')
            plt.imshow(img)
       
    def myfindcontours(self,img,mode):  
        #img=img.copy()
#         hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)       
        hsv=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)       
        lower_red = np.array([0, 40, 90])  #5, 90, 70  4, 65, 0
        upper_red = np.array([9, 160, 255]) #9, 160, 255  8, 155, 255
        mask1=cv2.inRange(hsv,lower_red,upper_red)

        lower_red2 = np.array([160, 40, 90])  #5, 90, 70
        upper_red2 = np.array([180, 160, 255]) #9, 160, 255
        mask2=cv2.inRange(hsv,lower_red2,upper_red2)

        mask=mask1+mask2
        #mask=mask1
        binary=cv2.threshold(mask,127,255,cv2.THRESH_BINARY)[1]
        #binary=cv2.dilate(binary,None,iterations=1)
        binary=cv2.dilate(binary,None,iterations=2)
        binary=cv2.erode(binary,None,iterations=2)        
    
        contours, _ = cv2.findContours(binary.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
               
        thr=img.size/3/100

        filter_contours=[]
        for cc in contours:
            if(cv2.contourArea(cc)>thr):
                filter_contours.append(cc)             
              #  print('contours area: {}'.format(area)) 
        
        sorted_cnts=sorted(filter_contours,key=cv2.contourArea,reverse=True) 
        try:
            cnt=sorted_cnts[0]
        except:
            return sorted_cnts
        else:
            if mode=='max':       
                hull = cv2.convexHull(cnt)
        #             platearea = cv2.contourArea(hull)#计算面积   
                return hull

            elif mode=='all':
                return sorted_cnts 
            

    def findarea(self,contours,imgsize):
        if len(contours)==0:
            area=imgsize
            contours_area=[0]
        else:
            contours_area=[]
            for cnt in contours:
                contours_area.append(cv2.contourArea(cnt))

            contours_area=sorted(contours_area)    

            sumarea=sum(contours_area)
            if sumarea>imgsize:
                area=contours_area[-2]
            else:
                area=imgsize-sumarea
        
        print('contours area: {}'.format(contours_area))
        print('food area: {}'.format(area))         
        return area
    
    def compute_iou(self,rec1, rec2):
        """
        computing IoU
        :param rec1: (y0, x0, y1, x1), which reflects
                (top, left, bottom, right)
        :param rec2: (y0, x0, y1, x1)
        :return: scala value of IoU
        """
        # computing area of each rectangles
        S_rec1 = (rec1[2] - rec1[0]) * (rec1[3] - rec1[1])
        S_rec2 = (rec2[2] - rec2[0]) * (rec2[3] - rec2[1])

        # computing the sum_area
        sum_area = S_rec1 + S_rec2

        # find the each edge of intersect rectangle
        left_line = max(rec1[1], rec2[1])
        right_line = min(rec1[3], rec2[3])
        top_line = max(rec1[0], rec2[0])
        bottom_line = min(rec1[2], rec2[2])

        # judge if there is an intersect
        if left_line >= right_line or top_line >= bottom_line:
            return 0
        else:
            intersect = (right_line - left_line) * (bottom_line - top_line)
            return intersect / (sum_area - intersect)
       
    def outinfo(self,base64str):                      
        #image = read_image_bgr(img_path)
        image=self.base64_to_image(base64str)      

        # copy to draw on
        draw = image.copy()
        draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)
        forcrop = draw.copy()

        # preprocess image for network
        image = preprocess_image(image)
        image, scale = resize_image(image)

        # foodplate contour & area
        platecontour = self.myfindcontours(forcrop,'max') 
       
        if len(platecontour) == 0:
            platearea= 0                  
        else: 
#             hull = cv2.convexHull(cnt)
#             platearea = cv2.contourArea(hull)#计算面积    
            platearea = cv2.contourArea(platecontour)
            self.plotsubfig(cv2.drawContours(forcrop.copy(),[platecontour], -1, (255,0,0), 5),2,1,2,1,8)
        #print('platearea: ',platearea)
       
        # process image        
        #start = time.time()
        with graph.as_default():
            boxes, scores, labels = self.model.predict_on_batch(np.expand_dims(image, axis=0))
        #print("processing time: ", time.time() - start)
        
        # correct for image scale
        boxes /= scale
         
        # visualize detections       
        scorelist,blist,namelist,foodarealist=[],[],[],[]
        n=1
        for box, score, label in zip(boxes[0], scores[0], labels[0]):
            # scores are sorted so we can break
            if score < 0.9:
                break                                   
            name=self.labelsname[label]
            color = label_color(label)
            b = box.astype(int)
            #print(b)

            # 截圖bbox取輪廓 返回食物面積(pixel)
            cropImg = forcrop[b[1]:b[3],b[0]:b[2]] 
            imghw= cropImg.size/3
            contours= self.myfindcontours(cropImg,'all') 
            foodarea = self.findarea(contours,imghw)     
            self.plotsubfig(cv2.drawContours(cropImg.copy(),contours, -1, (255,0,0), 2),1,2,4, n,15)           
            n+=1 
            
            if blist==[]:
                scorelist.append(score.item())
                blist.append(b.tolist())
                namelist.append(name)   
                foodarealist.append(foodarea)
                #畫框框
                self.draw_box2(draw, b, color=color)
                #畫字幕
                caption = "{} {:.3f}".format(name, score)
                draw=self.draw_caption2(draw, b, caption)
            else:
                for bb in blist:   
                    #print(box)
                    if self.compute_iou(bb,b)>0.5:
                        break
                    elif bb==blist[-1]:
                        scorelist.append(score.item())
                        blist.append(b.tolist())
                        namelist.append(name)   
                        foodarealist.append(foodarea)
                        #畫框框
                        self.draw_box2(draw, b, color=color)
                        #畫字幕
                        caption = "{} {:.3f}".format(name, score)
                        draw=self.draw_caption2(draw, b, caption)
                                                    
        success={
         "score":scorelist,
         "box":blist,
         "name":namelist,
         "foodarea":foodarealist,
         "platearea":platearea
        }
        #print(success)

        
        self.plotsubfig(draw,2,1,2,2,8)
        plt.show() 
          
        return success
```

```python
app = Flask(__name__)
tt=testmodel()

@app.route('/upload', methods=['POST'])
def post():
    #print(request.is_json)
    
    content = request.get_json()
    #print(content['photo'])
   
    #save image
    imgdata=base64.b64decode(content['photo'])
    filename=strftime("%Y%m%d%H%M%S")+'.png'
    filepath=os.path.join('/dataDisk/myfloder/jupyter/server/img/'+filename)   
    with open(filepath,'wb') as f:
        f.write(imgdata)
        
    # predict
    out=tt.outinfo(content['photo'])
    print(out)
    
    return jsonify(out)

app.run(host='140.116.67.155', port=5000)
```

```python
class testmodel:   
    def __init__(self):
#         self.model_path= os.path.join('..', 'snapshots', 'resnet50_csv_trans.h5')
        self.model_path= '/dataDisk/myfloder/jupyter/snapshots/0704.h5'
        self.csv_path='/dataDisk/myfloder/inputdata/csv/classes_title.csv'
        self.isplot=True     
        self.model=self.loadmodell()     
        self.labelsname=self.csvlabell()
        
    def loadmodell(self):
        # load retinanet model
        global model
        model = models.load_model(self.model_path, backbone_name='resnet50')
        global graph
        graph=tf.get_default_graph()
        return model
      
    def csvlabell(self):
        with open(self.csv_path, newline='') as csvfile: 
          # 讀取 CSV 檔內容，將每一列轉成一個 dictionary
            reader = csv.DictReader(csvfile)
            labels_to_names = {}
          # 以迴圈輸出指定欄位
            for row in reader:
                labels_to_names[int(row['label'])]= row['name']
          #print(labels_to_names)
        return(labels_to_names)  
    
#     def stringToRGB(self,base64_string):
#         #imgdata = base64.b64decode(str(base64_string))
#         imgdata = base64.b64decode(str(base64_string))
#         image = Image.open(io.BytesIO(imgdata))
#         return cv2.cvtColor(np.array(image), cv2.COLOR_BGR2RGB)
    
#     def base64_cv2(self,base64_str):
#         imgString = base64.b64decode(base64_str)
#         nparr = np.fromstring(imgString,np.uint8)  
#         image = cv2.imdecode(nparr,cv2.IMREAD_COLOR)
#         return image
    def base64_to_image(self,base64_code): 
        # base64解码
        img_data = base64.b64decode(base64_code)
        # 转换为np数组
        img_array = np.fromstring(img_data, np.uint8)
        # 转换成opencv可用格式
        img = cv2.imdecode(img_array, cv2.COLOR_RGB2BGR) 
        return img

    def draw_box2(self,image, box, color): 
        if self.isplot:
            draw_box(image, box, color=color)
        
    def draw_caption2(self,image,box,caption):
        if self.isplot:
            # 字體設定
            fontPath = "/usr/local/lib/python3.5/dist-packages/matplotlib/mpl-data/fonts/ttf/msj.ttf"
            font = ImageFont.truetype(fontPath,35)
            # 將 NumPy 陣列轉為 PIL 影像
            imgPil = Image.fromarray(image)  
            # 在圖片上加入文字
            draw = ImageDraw.Draw(imgPil)
            draw.text((box[0], box[1] - 10), caption, font = font, fill = (255,255,255))
            # 將 PIL 影像轉回 NumPy 陣列
            image = np.array(imgPil)
        return image
    
    def plotsubfig(self,img,fig,x,y,n,size): 
        if self.isplot:
            plt.figure(fig,figsize=(size,size))
            plt.subplot(x,y,n) # 2,4
            plt.axis('off')
            plt.imshow(img)

    def plotfig(self,img,fig,size): 
        if self.isplot:
            plt.figure(fig,figsize=(size,size))
            plt.axis('off')
            plt.imshow(img)
       
    def myfindcontours(self,img):  
        #img=img.copy()
#         hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)       
        hsv=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)       
        lower_red = np.array([0, 40, 90])  #5, 90, 70  4, 65, 0
        upper_red = np.array([9, 160, 255]) #9, 160, 255  8, 155, 255
        mask1=cv2.inRange(hsv,lower_red,upper_red)

        lower_red2 = np.array([160, 40, 90])  #5, 90, 70
        upper_red2 = np.array([180, 160, 255]) #9, 160, 255
        mask2=cv2.inRange(hsv,lower_red2,upper_red2)

        mask=mask1+mask2
        #mask=mask1
        binary=cv2.threshold(mask,127,255,cv2.THRESH_BINARY)[1]
        #binary=cv2.dilate(binary,None,iterations=1)
        binary=cv2.dilate(binary,None,iterations=2)
        binary=cv2.erode(binary,None,iterations=2)        
    
        contours, _ = cv2.findContours(binary.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
               
        thr=img.size/3/100

        filter_contours=[]
        for cc in contours:
            if(cv2.contourArea(cc)>thr):
                filter_contours.append(cc)             
              #  print('contours area: {}'.format(area))        
        return filter_contours

    def foodarea(self,imgsize,contours):
        contours_area=[]
        for cnt in contours:
            contours_area.append(cv2.contourArea(cnt))
            
        contours_area=sorted(contours_area)    
                    
        sumarea=sum(contours_area)
        if sumarea>imgsize:
            foodarea=contours_area[-2]
        else:
            foodarea=imgsize-sumarea
        
        print('contours area: {}'.format(contours_area))
        print('food area: {}'.format(foodarea))         
        return foodarea
    
    def compute_iou(self,rec1, rec2):
        """
        computing IoU
        :param rec1: (y0, x0, y1, x1), which reflects
                (top, left, bottom, right)
        :param rec2: (y0, x0, y1, x1)
        :return: scala value of IoU
        """
        # computing area of each rectangles
        S_rec1 = (rec1[2] - rec1[0]) * (rec1[3] - rec1[1])
        S_rec2 = (rec2[2] - rec2[0]) * (rec2[3] - rec2[1])

        # computing the sum_area
        sum_area = S_rec1 + S_rec2

        # find the each edge of intersect rectangle
        left_line = max(rec1[1], rec2[1])
        right_line = min(rec1[3], rec2[3])
        top_line = max(rec1[0], rec2[0])
        bottom_line = min(rec1[2], rec2[2])

        # judge if there is an intersect
        if left_line >= right_line or top_line >= bottom_line:
            return 0
        else:
            intersect = (right_line - left_line) * (bottom_line - top_line)
            return intersect / (sum_area - intersect)
       
    def outinfo(self,base64str):                      
        #image = read_image_bgr(img_path)
        image=self.base64_to_image(base64str)      

        # copy to draw on
        draw = image.copy()
        draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)
        forcrop = draw.copy()

        # preprocess image for network
        image = preprocess_image(image)
        image, scale = resize_image(image)

        # foodplate contour & area
        cropimghxw = forcrop.size/3
        platecontour = self.myfindcontours(forcrop)   
        sorted_cnts=sorted(platecontour,key=cv2.contourArea,reverse=True)
        cnt=sorted_cnts[0]

        if len(cnt) == 0:
            platearea=cropimghxw                  
        else: 
    #             hull = cv2.convexHull(cnt)
    #             platearea = cv2.contourArea(hull)#计算面积    
            platearea = cv2.contourArea(cnt)
        print(platearea)
       
        # process image        
        #start = time.time()
        with graph.as_default():
            boxes, scores, labels = self.model.predict_on_batch(np.expand_dims(image, axis=0))
        #print("processing time: ", time.time() - start)
        
        # correct for image scale
        boxes /= scale
         
        # visualize detections       
        scorelist,blist,namelist,foodarealist=[],[],[],[]
        n=1
        for box, score, label in zip(boxes[0], scores[0], labels[0]):
            # scores are sorted so we can break
            if score < 0.9:
                break                                   
            name=self.labelsname[label]
            color = label_color(label)
            b = box.astype(int)
            #print(b)

            # 截圖bbox取輪廓 返回食物面積(pixel)
            cropImg = forcrop[b[1]-10:b[3]+10,b[0]-10:b[2]+10] 
            imghxw= cropImg.size/3
            contours= self.myfindcontours(cropImg) 
            foodarea = self.foodarea(imghxw,contours)
            self.plotsubfig(cv2.drawContours(cropImg.copy(),contours, -1, (255,0,0), 2),1,2,4, n,15)           
            n+=1 
            
            if blist==[]:
                scorelist.append(score.item())
                blist.append(b.tolist())
                namelist.append(name)   
                foodarealist.append(foodarea)
                #畫框框
                self.draw_box2(draw, b, color=color)
                #畫字幕
                caption = "{} {:.3f}".format(name, score)
                draw=self.draw_caption2(draw, b, caption)
            else:
                for bb in blist:   
                    #print(box)
                    if self.compute_iou(bb,b)>0.5:
                        break
                    elif bb==blist[-1]:
                        scorelist.append(score.item())
                        blist.append(b.tolist())
                        namelist.append(name)   
                        foodarealist.append(foodarea)
                        #畫框框
                        self.draw_box2(draw, b, color=color)
                        #畫字幕
                        caption = "{} {:.3f}".format(name, score)
                        draw=self.draw_caption2(draw, b, caption)
                                                    
        success={
         "score":scorelist,
         "box":blist,
         "name":namelist,
         "foodarea":foodarealist,
         "platearea":platearea
        }
        #print(success)

        self.plotsubfig(cv2.drawContours(forcrop.copy(),platecontour, -1, (255,0,0), 5),2,1,2,1,8)
        self.plotsubfig(draw,2,1,2,2,8)
        plt.show() 
          
        return success
```
