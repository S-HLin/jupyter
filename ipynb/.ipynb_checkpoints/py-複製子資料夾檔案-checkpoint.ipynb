### 照片轉移資料夾

```python
import shutil
from os import listdir, walk
from os.path import isfile, isdir, join
```

```python
main_path = '/home/user/Downloads/food/未命名資料夾/train02/'
```

```python
for root, _, files in walk(main_path):
    print(root)
    for f in files:
        filepath = join(root, f)
        shutil.copy(filepath, main_path)
        print(filepath)
```
