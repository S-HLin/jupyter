### 照片轉移資料夾

```python
import shutil
from os import listdir, walk
from os.path import isfile, isdir, join
```

```python
main_path = '/dataDisk/Downloads/彙整5700~/'
```

```python
for root, _, files in walk(main_path):
    print(root)
    for f in files:
        filepath = join(root, f)
        shutil.copy(filepath, main_path)
        print(filepath)
```
