```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from os import listdir
from os.path import isfile, isdir, join

# 指定要列出所有檔案的目錄
#mypath = "/dataDisk/myfloder/inputdata/餐盤全盤/1113-2"
mypath = "/dataDisk/myfloder/jupyter/img/"
# 取得所有檔案與子目錄名稱
files = listdir(mypath)
y=7
x=len(files)//y+1
n=1
for f in files:
  # 產生檔案的絕對路徑
    fullpath = join(mypath, f)
    img=cv2.imread(fullpath)
    
    plt.figure(1)
    plt.subplot(x,y,n)
    plt.axis('off')
    plt.imshow(img)
    #plt.show()
    
    n+=1
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from os import listdir
from os.path import isfile, isdir, join

# 指定要列出所有檔案的目錄
#mypath = "/dataDisk/myfloder/inputdata/餐盤全盤/1113-2"
mypath = "/dataDisk/myfloder/jupyter/img/"
# 取得所有檔案與子目錄名稱
files = listdir(mypath)
n=1
for f in files:
  # 產生檔案的絕對路徑
  fullpath = join(mypath, f)
  img=cv2.imread(fullpath)
 
  plt.figure(1)
  plt.subplot(5,4,n)
  plt.axis('off')
  plt.imshow(img)
  
  n+=1
```
